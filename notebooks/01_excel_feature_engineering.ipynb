### one hot encoding for our featuer

```python
"""
note after seein the value counts of out featuer, there are a huge number of unique frequncy and combinations of possible values 
so using one hot encoding by pd.get_dummies wil not work. we have to make it manually
"""   

## possible values from sended doc:

gestalt_values = [
    "صعوبات الإغلاق",
    "صعوبات التقاطع",
    "التغير في الانحناءات",
    "التغير في حجم الزوايا",
    "تدوير محور الرسم",
    "تبسيط الرسم",
    "تجزئة الجشطالت",
    "الإضافة",
    "صعوبة رسم الأشكال المتداخلة"
]

notes_values = [
    "الرسم من أسفل لأعلى",
    "استخدام أكثر من اتجاه",
    "المداومة",
    "إعادة رسم الأجزاء (خطوط ثقيلة)",
    "اندفاع وسرعة في الرسم",
    "رسم فواصل بين الأشكال",
    "رسم كاريكاتيري"
]
```

```python

def normalize_text(text):
    text = str(text).strip() ## handle the outer spaces 
    
    ## handle the inner spaces and words mistaks
    replacements = {
        "،": ",",
        "-": ",",
        "صعوبات تقاطع": "صعوبات التقاطع",
        "صعوبات الإغلاق ": "صعوبات الإغلاق",
        "التغير في الإنحناءات": "التغير في الانحناءات",
        "تبسيط,": "تبسيط الرسم,",
        "تبسيط ": "تبسيط الرسم",
        "مداومة": "المداومة",
    }
    
    for old, new in replacements.items():
        text = text.replace(old, new)
    
    return text
```

```python
## main one hot encoded func, suitable for any column ->  التغيرات في الجشطالت ,ملاحظات دالة بشكل عام
def create_multi_value_features(df, column_name, possible_values, prefix):
    result = pd.DataFrame(index=df.index)
    
    normalized_col = df[column_name].fillna("لا يوجد").apply(normalize_text) ## normalize the column first to avoid mistaks of words
    
    for value in possible_values:
        result[f"{prefix}_{value}"] = normalized_col.apply(
            lambda x: 1 if value in x else 0
        )
    
    return result
```

```python
gestalt_features = create_multi_value_features(
    data,
    "التغيرات في الجشطالت",
    gestalt_values,
    "gestalt"
)

notes_features = create_multi_value_features(
    data,
    "ملاحظات دالة بشكل عام",
    notes_values,
    "note"
)

print(gestalt_features.shape)
print(notes_features.shape)

gestalt_features.head()
```

```python
single_value_cols = [
    "النوع",
    "الصف الدراسي",
    "العمر",
    "التسلسل",
    "موضع الرسم",
    "المسافة بين رسمين",
    "الاتصال أو التصادم",
    "استخدام الهامش",
    "تغيير موضع الورقة"
]

target_col = "التصنيفات المرضية"

basic_data = data[single_value_cols].copy()
y = data[target_col].copy()

X_basic = pd.get_dummies(basic_data, drop_first=False)

print("Basic features:", X_basic.shape)
print("Target:")
print(y.value_counts())
```
